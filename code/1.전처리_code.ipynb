## 인구정보 + 건물연면적

```python
#C:\Users\Administrator\Downloads\SBJ_2012_002
df3 = gpd.read_file('input/3.용인시_인구정보(총인구수)_격자.geojson', encoding='cp949')
df3.loc[df3['val'].apply(pd.isna), 'val'] = 0
df3.columns=['gid','총인구수','geometry']
df3.head()
```

```python
df4=gpd.read_file('input/4.용인시_인구정보(고령)_격자.geojson')
df4.loc[df4['val'].apply(pd.isna), 'val'] = 0
df4.columns=['gid','고령인구수','geometry']
df4.head()
```

```python
df5= gpd.read_file('input/5.용인시_인구정보(생산가능)_격자.geojson', encoding='cp949')
df5.loc[df5['val'].apply(pd.isna), 'val'] = 0
df5.columns=['gid','생산가능인구수','geometry']
df5.head()
```

```python
df6= gpd.read_file('input/6.용인시_인구정보(유소년)_격자.geojson', encoding='cp949')
df6.loc[df6['val'].apply(pd.isna), 'val'] = 0
df6.columns=['gid','유소년인구수','geometry']
df6.head()
```

```python
df_p=df3.merge(df4, on='gid')
df_p.head()
```

```python
df_p=df_p.merge(df5,on='gid')
df_p=df_p.merge(df6,on='gid')
```

```python
df_p.head()
```

```python
del df_p["geometry_y"]
df_p.columns=['gid','all_cnt','geometry','old_cnt','adult_cnt','geo','young_cnt']
del df_p["geo"]
```

```python
df_p.head()
```

```python
df10= gpd.read_file('input/10.용인시_건물연면적_격자.geojson', encoding='cp949')
#df10=gpd.read_file(input_path.joinpath('10.용인시_건물연면적_격자.geojson'))
df10.loc[df10['val'].apply(pd.isna), 'val'] = 0
df10.columns=['gid','건물연면적','geometry']
df10.head()
```

```python
df_p=df_p.merge(df10,on='gid')
del df_p['geometry_y']
df_p.columns=['gid','all_cnt','geometry','old_cnt','adult_cnt','young_cnt','건물연면적']
```

```python
df_p.head()#df_p는 3456 +10 병합
```

### 500 격자로 통합

```python
grid100 = pd.read_csv('grid100to500.csv')
grid100.head()
```

```python
temp = pd.merge(df_p,grid100,left_on='gid',right_on='gid100')
fin_df_p = temp.groupby('gid500').sum().reset_index()
fin_df_p
```

## 소상공인 매출

```python
df16 = gpd.read_file('input/16.용인시_소상공인_매출정보.geojson')
# df17 = pd.read_csv('input/17.용인시_소상공인_매출정보.csv')
df16 = df16.fillna(0)
df16.head(n=3)
```

```python
df16.iloc[:,45:-1] = df16.iloc[:,45:-1].replace('Y',1)
df16.iloc[:,45:-1].sum()
```

### 500 격자로 통합

```python
grid250 = pd.read_csv('grid250to500.csv')
grid250.head()
```

```python
temp2 = pd.merge(df16,grid250,left_on='gid',right_on='gid250')
temp2
fin_df_p2 = temp2.groupby('gid500').sum().reset_index()
fin_df_p2.head(3)
```

```python
len(fin_df_p2)
```

## 독립변수

```python
df_val = pd.read_csv('대격자별_독립변수.csv')
df_geo = gpd.read_file('500grid.shp')
df_geo = df_geo[['gid','geometry']]
df_geo.columns = ['gid500','geometry']
df_val.head()
```

## 전체 데이터 통합

```python
temp3 = pd.merge(fin_df_p,fin_df_p2,on='gid500')
temp3 = pd.merge(temp3,df_val,on='gid500')
fin_df = pd.merge(temp3,df_geo,on='gid500')
fin_df.head()
```

```python
# fin_df.to_csv('fin_df.csv',index=False)
```

## 최종 데이터 geojson 파일로 저장

```python
fin_gdf=gpd.GeoDataFrame(fin_df)
fin_gdf.to_file('fin_df.geojson', driver='GeoJSON')
```
